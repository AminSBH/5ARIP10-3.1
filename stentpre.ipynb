```python
import os
import warnings

import numpy as np

warnings.filterwarnings('ignore')
np.random.seed(123)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from autogluon.multimodal import MultiModalPredictor
```

```python
path = 'Dataset.xlsx'
data = pd.read_excel(path)
label = 'Vessels with ISR'
data
```

```python
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
predictor = MultiModalPredictor(label=label, problem_type="multiclass")
class_weights = {'0': 1.0, '1': 3.0, '2': 3.0, '3': 3.0}
predictor.fit(
    hyperparameters={
        "env.num_gpus": 1,
        "optimization.loss_function": "focal_loss",
        "optimization.focal_loss.gamma": 1.0,
        "optimization.focal_loss.reduction": "sum",
        "optimization.max_epochs": 10,
        'class_weights': class_weights
    },
    train_data=train_data,
) 

predictor.evaluate(test_data, metrics=['accuracy'])
```

```python
results = predictor.fit_summary()
print(results)
```

```python
y_true = test_data[label]  # true label
y_pred = predictor.predict(test_data)  # predicted label
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_true, y_pred)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy_score(y_true, y_pred)
```

```python
recall_score(y_true, y_pred, average="weighted")
```

```python
precision_score(y_true, y_pred, average="weighted")
```

```python
f1_score(y_true, y_pred, average="weighted")
```
